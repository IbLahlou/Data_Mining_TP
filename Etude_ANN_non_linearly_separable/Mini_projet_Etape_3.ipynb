```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres des sinusoïdes
amplitude = 1
frequence = 1
dephasage = np.pi / 2  # Déphasage de 90 degrés (en radians)

# Temps
t = np.linspace(0, 7*np.pi/4, 500)  # Nombre total de points à 500




# Génération des données pour les deux classes
classe1 = amplitude * np.sin(frequence * t)
classe2 = amplitude * np.sin(frequence * t + dephasage)

# Ajout de bruit aux données
bruit1 = 0.2 * np.random.randn(500)  # Génération de bruit gaussien pour la première moitié
bruit2 = 0.2 * np.random.randn(500)  # Génération de bruit gaussien pour la deuxième moitié
classe1 += bruit1
classe2 += bruit2

# Tracé des nuages de points pour chaque moitié
plt.scatter(t, classe1, label='Classe 1')
plt.scatter(t, classe2, label='Classe 2')
plt.xlabel('Temps')
plt.ylabel('Amplitude')
plt.title('Nuages de points pour deux classes de données (moitiés)')
plt.legend()
plt.show()
```
